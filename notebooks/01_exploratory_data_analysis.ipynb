```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)
sns.set(style="whitegrid")
```

```python
df = pd.read_csv("../data/raw/weatherAus.csv")

df.head()
```

```python
plt.figure(figsize=(14, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis', yticklabels=False)
plt.title('Missing Values Map')
plt.show()
```

```python
df["RainTomorrow"].value_counts()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x="RainTomorrow", data=df, palette="Set2")
plt.title("Distribution of Rain Tomorrow Target")
plt.xlabel("Rain Tomorrow (No = Dry, Yes = Rainy)")
plt.ylabel("Count (Number of Days)")
plt.show()
```

```python
top_10 = df['Location'].value_counts().head(10).index
top_10_locations = df[df['Location'].isin(top_10)]

sns.countplot(x="Location", hue="RainTomorrow", data=top_10_locations, palette="viridis")
plt.title("Rain Tomorrow Distribution across Sample Locations")
plt.xlabel("Location")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["MaxTemp"].dropna(), bins=20, kde=True, color="tomato")
plt.title("Maximum Temperature Distribution")
plt.xlabel("Max Temperature (°C)")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(7, 4))
sns.countplot(x="RainToday", hue="RainTomorrow", data=df, palette="cool")
plt.title("Rain Today vs Rain Tomorrow")
plt.xlabel("Did it Rain Today?")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x="RainTomorrow", y="Humidity3pm", data=df, palette="Pastel1")
plt.title("3 PM Humidity Levels by Rain Tomorrow Outcome")
plt.xlabel("Will it Rain Tomorrow?")
plt.ylabel("Humidity at 3 PM (%)")
plt.show()
```

```python
# Temporarily Yes = 1 No = 0
df['RainTomorrow_Numeric'] = df['RainTomorrow'].map({'Yes': 1, 'No': 0})

# Heatmap only with numbers
numeric_df = df.select_dtypes(include=['float64', 'int64'])
corr = numeric_df.corr()

plt.figure(figsize=(14, 10))
sns.heatmap(corr, annot=False, cmap="coolwarm", linewidths=0.2)
plt.title("Weather Feature Correlation Heatmap (With Target)")
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df_temp = df.copy()

df_temp['Month'] = df_temp['Date'].dt.month
df_temp['RainTomorrow_Numeric'] = df_temp['RainTomorrow'].map({'No': 0, 'Yes': 1})

plt.figure(figsize=(9, 5))
sns.lineplot(x='Month', y='RainTomorrow_Numeric', data=df_temp, marker='o', color='teal', linewidth=2.5)
plt.title('Monthly Rain Probability Trend in Australia', fontsize=14)
plt.xlabel('Month of the Year')
plt.ylabel('Average Rain Probability')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# Calculate pressure change
df_temp['Pressure_Drop'] = df_temp['Pressure9am'] - df_temp['Pressure3pm']

plt.figure(figsize=(8, 5))
sns.kdeplot(data=df_temp, x='Pressure_Drop', hue='RainTomorrow', fill=True, common_norm=False, palette='crest', alpha=0.5)
plt.title('Distribution of Pressure Drop (9AM - 3PM) vs Rain Tomorrow', fontsize=12)
plt.xlabel('Pressure Drop (hPa)')
plt.ylabel('Density')
plt.show()
```
